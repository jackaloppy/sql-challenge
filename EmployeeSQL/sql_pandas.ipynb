```python
from sqlalchemy import create_engine
from postgres_info import username, password
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/sql-challenge')
connection = engine.connect()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
salary = pd.read_sql("SELECT salary FROM salaries", connection)
```

```python
plt.style.use('ggplot')
plt.hist(salary["salary"])
plt.title("Salary ranges frequency for all employees")
plt.ylabel("Count")
plt.xlabel("Salary ($)")
plt.grid(c='w')
plt.show()

#The salary range frequency histogram looks good so far.
```

```python
sql_query = "SELECT t.title, s.salary FROM titles t LEFT JOIN " \
            "employees e ON e.emp_title_id = t.title_id LEFT JOIN "\
            "salaries s ON s.emp_no = e.emp_no"
salary_title = pd.read_sql(sql_query, connection)
```

```python
salary_group = salary_title.groupby("title")["salary"].mean().round().astype(int)
```

```python
plt.style.use('ggplot')
plt.bar(salary_group.index, salary_group.values)
plt.grid(c='w')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Employee titles")
plt.ylabel("Average salary ($)")
plt.show()

## So the average staff salary is higher than manager. Obviously it's faked data" 
```
